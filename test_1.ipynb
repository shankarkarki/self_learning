```python
ds=pd.read_csv("stock_data.csv",header=1)
ds
```

here understand the concept of header which ever header number you chose from zero indexing will be considered as the first element header and other following the number of header are just elements components of it.

also, index starts from 0 and hence header=1 means rows#2 in the csv files

In the case you have a missing headers or you just want to change the headers of the csv files,firstly make headers none and
follow it by the name you just want to replace by.
Example::

```python
dl=pd.read_csv("stock_data.csv",header=None)
dl
```

Header name 0,1,2,3,4 dont sound professional just replace by any name.

```python
de=pd.read_csv("stock_data.csv",header=None,names=["srk","krk","snk","kns","krs"])
de
```

In the case you have a very big CSV files and only want to read few your desired rows.(excluding headers)

```python
dz=pd.read_csv("stock_data.csv",nrows=3)
dz
```

Here,first three rows has been only taken from the data frame having many number of rows. Please understand the concept 
and need for doing this.
